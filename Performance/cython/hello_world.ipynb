Test function:

```python
def test(x):
    y = 1
    for i in range(1, x+1):
        y *= i
    return y
```

Timing wrapper:

```python
from typing import Callable
import time

class TimeIt:

    def __init__(self, print_time: bool = True):
        self.print_time = print_time

    def __enter__(self):
        self.start_time = time.time()

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.end_time = time.time()
        self.running_time = self.end_time - self.start_time
        if self.print_time:
            print(f"The function took {self.running_time} s")
```

```python
with TimeIt() as t:
    result = [test(int(10)) for _ in range(int(1e6))]
```

```python
import run_cython

with TimeIt() as t:
    result = [run_cython.test(int(10))  for _ in range(int(1e6))]
```
